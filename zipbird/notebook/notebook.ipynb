```python
from datetime import date
import norgatedata
import pandas as pd
```

```python
def get_sp500_on_date(d):
    result = []
    all_sp500_symbols = norgatedata.watchlist_symbols('S&P 500')
    for symbol in all_sp500_symbols:
        idx = norgatedata.index_constituent_timeseries(symbol=symbol,
                                                       indexname='S&P 500',
                                                       start_date=d,
                                                       timeseriesformat='pandas-dataframe')
        if idx.iloc[0]['Index Constituent'] == 1:
            result.append(symbol)
    return result
```

```python
sp500_symbols = get_sp500_on_date('2021-01-01')
```

```python
def get_performance_in_date_range(symbols, start_date, end_date):
    result = []
    for symbol in symbols:
        p = norgatedata.price_timeseries(symbol=symbol,
                                        stock_price_adjustment_setting=norgatedata.StockPriceAdjustmentType.TOTALRETURN,
                                        padding_setting=norgatedata.PaddingType.ALLCALENDARDAYS,
                                        start_date=start_date,
                                        end_date=end_date,
                                        timeseriesformat='pandas-dataframe')
        result.append({'symbol': symbol,
                       'start_price': p.iloc[0]['Close'],
                       'start_date': p.index[0],
                       'end_price' :p.iloc[-1]['Close'],
                       'end_date': p.index[-1]
                       })
    return pd.DataFrame.from_dict(result)
```

```python
p_2021 = get_performance_in_date_range(sp500_symbols, '2021-01-01', '2021-12-31')
```

```python
p_2021['change'] = p_2021['end_price'] / p_2021['start_price'] - 1
```

```python
for _, row in p_2021.sort_values(['change']).iterrows():
    #print(row)
    print(f"{row['symbol']},{row['start_price']},{row['end_price']}, {row['change']*100:.2f}")
```
